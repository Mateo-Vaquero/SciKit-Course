```python
#Imports dataset from the sklearn toy datasets
from sklearn.datasets import load_iris
```

```python
#Loads the dataset as X and Y variables
X, y = load_iris(return_X_y=True)
```

```python
#Imports submodule KNeighbotRegressor  from sklearn neighbors module
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
```

```python
#Create variable for model and points to model object
model = KNeighborsRegressor()
#model2 = LinearRegression()
```

```python
#Fits the model with data
model.fit(X, y)
#model2.fit(X, y)
```

```python
#Utilizing model to predict X variable. Creates prediction variable, outputs array 
prediction = model.predict(X)
```

```python
#Imports the matplotlib library 
from matplotlib import pyplot as plt
```

```python
#Creates scatterplot. Predictions on X axis and predictive values on Y axis
plt.scatter(prediction , y)
```

```python
#Imports from preprocessing module, a scaler object for our data
from sklearn.preprocessing import StandardScaler
```

```python
#Imports from pipeline module, a pipeline class object. allows to  chain processing steps
from sklearn.pipeline import Pipeline
```

```python
#starts new pipeline object. Which needs a list of tuples. Which need a variable name with a Step object
pipe = Pipeline([
    ("scale", StandardScaler()),
    ("model", KNeighborsRegressor(n_neighbors=1))
])
```

```python
#Fits X and Y data with the new pipeline of processing steps 
pipe.fit(X, y)
```

```python
#Using our new model with a scaler to predict X values. Outputs array
predScaled = pipe.predict(X)
```

```python
#Creates scatterplot of new model with scaler. Predictions on X axis and predictive values on Y axis
plt.scatter(predScaled , y)
```
